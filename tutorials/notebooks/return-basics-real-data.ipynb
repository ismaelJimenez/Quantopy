# Risk Adjusted Returns with Real Data

## Retrieve Data for portfolio formed on ME

```python
import quantopy as qp
```

```python
ds = qp.get('Portfolios_Formed_on_ME', 'famafrench')
print(ds['DESCR'])
```

```python
portfolio = ds[1]
portfolio.head()
```

```python
columns = ['Lo 10', 'Hi 10']
returns = portfolio[columns]
returns.head()
```

```python
returns.plot(title='Stock Returns')
```

## Calculate Standard Deviation of returns

```python
returns.std()
```

## Calculate Mean fo the returns

```python
returns.mean()
```

## Annualization

## Annual Rate of Return

```python
returns.effect(qp.stats.period.MONTHLY)
```

## Annual Volatility

```python
returns.effect_vol(qp.stats.period.MONTHLY)
```
